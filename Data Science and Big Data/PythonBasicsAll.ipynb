## 2.6. Functions: Custom 

```python
def funcfoo1():
    print("this is printed inside the function funcfoo1")

funcfoo1()     
```

```python
def funcfoo2(input):
    
    print("this is printed inside the function funcfoo2: ", input + 10)
  
funcfoo2(100)
```

```python
# The below code produces an error since str cannot be added to int

def funcfoo2(input) :
    
    print("this is printed inside the function funcfoo2: ", input + 10)
  
funcfoo2("cat")
```

```python
# Let's fix the code to make the failure more elegant so it does not crash the program
# We use try-except blocks
# Any code that is likely to produce an error (if input is incorrect) is placed inside a try block
# in the except block, write what you want to do if there is an exception (error)

def funcfoo2(input):
    
    try:
        print("this is printed inside the function funcfoo2: ",      input + 10)
    except:
        print("can't do it")
  
funcfoo2("cat")
```

```python
def funcfoo3(s, t):
        
    print("this is printed inside the function funcfoo3")
    return s+t

# The above function has a return value, which is the sum s+t
# When we call the function, the return value is returned to the function call, 
# but since we didnt ask to print it, it will not be shown in the output (outside of Jupyter Notebooks)

funcfoo3(10,30)
```

```python
# Explicitly calling the print() function will ensure that the answer 40 is printed to the output (here, console)

print(funcfoo3(10,30))
```

```python
# If a functions defines a return value (see the keyword return above)
# you can save the answer returned in a new variable outside of the function when you call it 

ans = funcfoo3(10,30)
print(ans)

# Now this value that we saved is available for us to use outside of the function, anywhere in our program

print ("The answer from funcfoo3 times 3 is" , ans *3)
```

```python
def funcfoo4(param1,param2):
  print(param1, "is", param2, "years young.")

# the order of arguments matters

funcfoo4("mary", 40.5)

funcfoo4(40.5, "mary")
```

```python
def funcfoo5(s, t):
        
    # You can call other functions from another function
    
    funcfoo1()
    funcfoo2(s)
    print("this is printed inside the function funcfoo5")
    return s+t

funcfoo5(10,30)
```

```python
# Here we define one positional or required parameter (param1), and another keyword (optional) parameter param2
def funcfoo6(param1,param2=40):
    a = param1 > 50
    b = param2 > 50
    return (a, b)

ans6 = funcfoo6(20, 60)
print(ans6)
```

## 2.16 Control Flow: Loops

### for loop

Setting a loop counter or defining the iteration using the **in** operator with a **sequence**. The sequence can be a string, list, range, tuple etc. It is common to use range as the sequence if you want an explicit numeric counter (run 10 times)

```python
sample = "ABCDE"

# How many times will this loop run?
# using the in keyword will ensure that the loop runs as long as 
# there is another element available in the collection named sample
# here there are 5 characters (as string is made up of chars), so loop will execute 5 times

for varcounter in sample:
    print(varcounter)
```

For loop: counting with range

```python
for i in range(2, 10, 1):                                       
  print(i, end=' ')          
print("\n")
```

```python
for i in range(10):                                       
    if i % 2 ==0 :
        print(f"{i} is an even number")      #an f-string is used to replace the variable within {} with actual values
    else :
        print(f"{i} is an odd number")
```

### using continue and break within a loop

```python
count=0
for letter in sample:
    
    if (letter == 'B'):
            continue         #pass, go to the beginning of the loop, dont do anything else inside the loop on this run
    elif (letter == 'D'):
            break            #exit the loop now, dont do anything else inside the loop
    else :
        print(letter)
    count=count+1        

# Why is the answer 2? can you explain?

print("count is ", count)
```

#### sorted()

a top-level function that provides us with some flexible ways to sort. This function takes the data structure/collection to sort as input. By default, sorted will sort alphabetically by the value in the collection.
To change the default behavior, we can specify a parameter called **key**, where we can write code for a custom sort. Often we use lambda functions for this.

```python
# sort a list by the default sort

print(blist)
sorted(blist)
```

```python
# sort by the last character of the word in the list

sorted(blist, key = lambda x: x[-1])
```

```python
# let's sort a dict by specifying how in the parameter key
# we use dict.items() to obtain the elements of the dict as a key-value pair as a tuple (key, value)

print(capitalsdict)
sorted(capitalsdict.items())
```

```python
# sort using the length of the key. 
# then to access the key from the tuple we use index 0, here x[0]
# and to access the value from the tuple we use index 1, here x[1]

sorted(capitalsdict.items(), key=lambda x : len(x[0]))
```

```python
# sort by the length of the values

sorted(capitalsdict.items(), key=lambda x : len(x[1]))
```

```python
# sort by the length of the values

sorted(capitalsdict.items(), key=lambda x : len(x[1]), reverse= True)
```
