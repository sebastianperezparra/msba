## Task 13

```python
# 13a
inUSADF = flightsAllDF.filter((F.col('origin') != 'United States') & (F.col('dest') == 'United States')).groupBy('year').sum('numflights').orderBy(['sum(numflights)'], ascending=[True]).select('year',F.col('sum(numflights)').alias('totalflightsin'))

display(inUSADF)
```

```python
# 13b
outUSADF = flightsAllDF.filter((F.col('origin') == 'United States') & (F.col('dest') != 'United States')).groupBy('year').sum('numflights').orderBy(['sum(numflights)'], ascending=[True]).select(F.col('sum(numflights)').alias('totalflightsout'),'year')

USADF = inUSADF.join(outUSADF, on='year', how='inner')

USADF.select('*',(USADF.totalflightsin > USADF.totalflightsout).alias('morein')).orderBy(['year'], ascending=[True]).show()
```
