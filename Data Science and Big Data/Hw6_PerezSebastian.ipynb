```python
%sql



CREATE CATALOG IF NOT EXISTS is6493;

CREATE SCHEMA IF NOT EXISTS is6493.homework;

CREATE VOLUME IF NOT EXISTS is6493.homework.hw6;
```

```python
# define a base path



path_to_hw6 = "/Volumes/is6493/homework/hw6"  

# this function returns the newly created folder's path



def createtaskfolder (current_task):

   path_to_checkpoint = path_to_hw6 + "/" + current_task

   dbutils.fs.mkdirs(path_to_checkpoint)

   return path_to_checkpoint

def deletecheckpoints(current_task):



    path_to_checkpointoffsets = "/Volumes/is6493/homework/hw6/" + current_task 



    try:

        dbutils.fs.ls(path_to_checkpointoffsets)

        folder_exists = True

    except Exception as e:

        if "No such file or directory" in str(e):

            folder_exists = False

        else:

            raise e



    if folder_exists:

        dbutils.fs.rm(path_to_checkpointoffsets, recurse=True)

        print(f"Folder '{current_task}' deleted from volume '{path_to_hw6}'.")

    else:

        print(f"No action needed. Folder '{path_to_checkpointoffsets}' not found.")
```

```python
# 6a
streamingDF = spark.readStream.schema(dataSchema).option("maxFilesPerTrigger", 1).json("dbfs:/databricks-datasets/definitive-guide/data/activity-data/*.json")

#streamingDF = spark.readStream.option("maxFilesPerTrigger", 1).json("dbfs:/databricks-datasets/definitive-guide/data/activity-data/*.json")
```

## Task 13

```python
# 13a

path_to_checkpoint = createtaskfolder("task13")

historicalAvgDF = staticDF.groupBy('gt', 'Model').agg(F.avg('x').alias('x_avg_h'), F.avg('y').alias('y_avg_h'), F.avg('z').alias('z_avg_h'))
```

```python
# 13b
streamingAvgDF = streamingDF.groupBy('gt', 'Model').agg(F.avg('x').alias('x_avg_s'), F.avg('y').alias('y_avg_s'), F.avg('z').alias('z_avg_s'))

joinedDF = streamingAvgDF.join(historicalAvgDF, on=['gt', 'Model'], how='inner')

streamJoinQuery = joinedDF.writeStream.queryName('JoinedData').format("memory").outputMode("complete").trigger(availableNow=True).option('checkpointLocation', path_to_checkpoint).start()
```

```python
# 13c
for _ in range(4):
    display(spark.sql("SELECT * FROM JoinedData"))
    sleep(8)
```

```python
# 13d
streamJoinQuery.stop()

path_to_checkpointoffsets = "/Volumes/is6493/homework/hw6/task13"

dbutils.fs.rm(path_to_checkpointoffsets, recurse=True)
```
